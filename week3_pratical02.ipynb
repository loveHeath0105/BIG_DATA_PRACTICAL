**What is the geometry value for the street named 'Atlantic Commons'?**

```python
con.sql("""
SELECT ST_AsText(geom), ST_GeometryType(geom)
FROM nyc_streets
WHERE name = 'Atlantic Commons'
""")
```

**What neighborhood and borough is Atlantic Commons in?**

```python
con.sql("""
SELECT boroname, name, ST_GeometryType(geom)
FROM nyc_neighborhoods
WHERE ST_Intersects(
  geom, ST_GeomFromText('LineString(586781.7015777241 4504202.153143394, 586863.5196448397 4504215.988170098)')

)
""")
```

**What streets does Atlantic Commons join with?**

```python
con.sql("""
SELECT name, ST_GeometryType(geom)
FROM nyc_streets
WHERE ST_DWithin(
  geom, 'LINESTRING(586781.7015777241 4504202.153143394, 586863.5196448397 4504215.988170098)', 0.1
)
""")
```

```python
con.sql("""
SELECT name, ST_GeometryType(geom)
FROM nyc_streets
WHERE ST_Touches(
  geom, ST_GeomFromText('LINESTRING(586781.7015777241 4504202.153143394, 586863.5196448397 4504215.988170098)')
)
""")
```

**Approximately how many people live on (within 50 meters of) Atlantic Commons?**

```python
con.sql("""
SELECT sum(POPN_TOTAL)
FROM nyc_census_blocks
WHERE ST_DWithin(
  geom, ST_GeomFromText('LineString(586782 4504202, 586864 4504216)'), 50)
""")
```
